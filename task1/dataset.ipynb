```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set base directory (your current working directory)
base_dir = r"C:\Users\athar\OneDrive\Desktop\Atharva\Github\open_source\humandArt\task1"

# Construct file paths dynamically
artist_csv = os.path.join(base_dir, "WikiArt Dataset", "Artist", "artist_train")  # No .csv extension
genre_csv = os.path.join(base_dir, "WikiArt Dataset", "Genre", "genre_train.csv")
style_csv = os.path.join(base_dir, "WikiArt Dataset", "Style", "style_train.csv")

# Check if files exist
for path in [artist_csv, genre_csv, style_csv]:
    print(f"Checking: {path} -> Exists: {os.path.exists(path)}")

# Load CSVs
df_artist = pd.read_csv(artist_csv)  # No .csv extension
df_genre = pd.read_csv(genre_csv)
df_style = pd.read_csv(style_csv, header=None, names=["Path", "Class"])  # Assuming no header in style CSV

# Extract class counts
artist_classes = df_artist.iloc[:, -1].value_counts().sort_index()
genre_classes = df_genre.iloc[:, -1].value_counts().sort_index()
style_classes = df_style["Class"].value_counts().sort_index()

# Function to plot class distribution
def plot_class_distribution(class_counts, title):
    plt.figure(figsize=(12, 6))
    sns.barplot(x=class_counts.index, y=class_counts.values, palette="viridis")
    plt.xlabel("Class")
    plt.ylabel("Total Count")
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()

# Plot each dataset separately
plot_class_distribution(artist_classes, "Class Distribution in Artist Dataset")
plot_class_distribution(genre_classes, "Class Distribution in Genre Dataset")
plot_class_distribution(style_classes, "Class Distribution in Style Dataset")

print("✅ Data loaded and graphs generated successfully!")
```
